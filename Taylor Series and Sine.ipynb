```python
#import statements 

import numpy as np
import math as m 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 100) # (Lower bound, upper bound, number of points)
y = np.sin( x )
```

```python
plt.plot(x,y)
```

```python
def my_sin(x, n):
    y_out = np.zeros(np.shape(x))
    for i in range(n):
        y_out = y_out + (-1)**i * np.divide(np.power(x,(2*i+1)) , m.factorial(2*i+1))

    return y_out
```

```python
p_1 = my_sin(x, 1)

plt.plot(x,y)
plt.plot(x,p_1)
```
